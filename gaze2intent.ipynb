```python
# Enabling Left click - Right click - Double Click
import cv2
import mediapipe as mp
import pyautogui
import numpy as np
import time

# Initialize MediaPipe
mp_face_mesh = mp.solutions.face_mesh
mp_drawing = mp.solutions.drawing_utils
face_mesh = mp_face_mesh.FaceMesh(
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

cap = cv2.VideoCapture(0)

# Custom drawing specs
eye_landmark_drawing_spec = mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=1, circle_radius=1)  # Unused
iris_drawing_spec = mp_drawing.DrawingSpec(color=(255, 255, 255), thickness=2, circle_radius=2)  # White for pupil
connection_drawing_spec = mp_drawing.DrawingSpec(color=(100, 100, 100), thickness=1, circle_radius=1)

# Movement Parameters
smoothing_factor = 0.6  # For smooth cursor movement
blink_threshold = 5.0  # Blink detection threshold
cursor_scale = 0.7  # Increased for faster cursor
blink_window = 0.5  # Time window for double blink detection (seconds)

# Screen dimensions
screen_w, screen_h = pyautogui.size()

# Iris and eye indices (MediaPipe 468 landmarks)
LEFT_IRIS = [474, 475, 476, 477]
RIGHT_IRIS = [469, 470, 471, 472]
LEFT_EYE = [33, 7, 163, 144, 145, 153, 154, 155, 133, 173, 157, 158, 159, 160, 161, 246]
RIGHT_EYE = [362, 382, 381, 380, 374, 373, 390, 249, 263, 466, 388, 387, 386, 385, 384, 398]

# For smooth cursor movement
prev_cursor_pos = None
cursor_smoothing = 0.6  # For responsive cursor

# For double blink detection
right_blink_times = []
both_blink_times = []

def get_iris_center(landmarks, indices, img_w, img_h):
    # Calculate precise iris center
    x = sum(landmarks[idx].x * img_w for idx in indices) / len(indices)
    y = sum(landmarks[idx].y * img_h for idx in indices) / len(indices)
    return x, y

def normalize_iris_position(landmarks, eye_indices, iris_indices, img_w, img_h):
    # Get iris center
    iris_x, iris_y = get_iris_center(landmarks, iris_indices, img_w, img_h)
    
    # Get eye bounding box (using min/max of eye landmarks)
    eye_x_coords = [landmarks[idx].x * img_w for idx in eye_indices]
    eye_y_coords = [landmarks[idx].y * img_h for idx in eye_indices]
    min_x, max_x = min(eye_x_coords), max(eye_x_coords)
    min_y, max_y = min(eye_y_coords), max(eye_y_coords)
    
    # Normalize iris position within eye bounding box with expanded range
    if max_x > min_x and max_y > min_y:
        norm_x = (iris_x - min_x) / (max_x - min_x)
        norm_y = (iris_y - min_y) / (max_y - min_y)
        # Expand vertical range to ensure full movement
        norm_y = norm_y * 1.2 - 0.1  # Scale and shift to cover more vertical range
        norm_y = max(0.0, min(1.0, norm_y))  # Clamp to [0, 1]
    else:
        norm_x, norm_y = 0.5, 0.5  # Fallback to center if eye bounds are invalid
    
    return norm_x, norm_y, iris_x, iris_y

def draw_eye_landmarks(frame, landmarks, eye_indices, iris_indices, img_w, img_h):
    # Draw white donut ring at iris center
    _, _, iris_x, iris_y = normalize_iris_position(landmarks, eye_indices, iris_indices, img_w, img_h)
    cv2.circle(frame, (int(iris_x), int(iris_y)), 5, (255, 255, 255), -1)  # Outer white circle
    cv2.circle(frame, (int(iris_x), int(iris_y)), 4, (0, 0, 0), -1)  # Inner black circle for donut effect
    
    return iris_x, iris_y

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Flip frame horizontally to correct mirror effect
    frame = cv2.flip(frame, 1)

    img_h, img_w = frame.shape[:2]
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(rgb_frame)

    if results.multi_face_landmarks:
        for face_landmarks in results.multi_face_landmarks:
            # Process both eyes
            left_norm_x, left_norm_y, left_iris_x, left_iris_y = normalize_iris_position(
                face_landmarks.landmark, LEFT_EYE, LEFT_IRIS, img_w, img_h)
            right_norm_x, right_norm_y, right_iris_x, right_iris_y = normalize_iris_position(
                face_landmarks.landmark, RIGHT_EYE, RIGHT_IRIS, img_w, img_h)
            
            # Draw white donut rings
            draw_eye_landmarks(frame, face_landmarks.landmark, LEFT_EYE, LEFT_IRIS, img_w, img_h)
            draw_eye_landmarks(frame, face_landmarks.landmark, RIGHT_EYE, RIGHT_IRIS, img_w, img_h)
            
            # Use average of normalized iris positions for cursor
            norm_x = (left_norm_x + right_norm_x) / 2
            norm_y = (left_norm_y + right_norm_y) / 2
            
            # Map normalized coordinates to screen, inverting y-axis
            cursor_x = int(norm_x * screen_w * cursor_scale)  # Scaled for horizontal
            cursor_y = int((1 - norm_y) * screen_h)  # Full range for vertical
            
            # Smooth cursor movement
            if prev_cursor_pos is None:
                prev_cursor_pos = (cursor_x, cursor_y)
            else:
                smooth_x = int(prev_cursor_pos[0] * cursor_smoothing + cursor_x * (1 - cursor_smoothing))
                smooth_y = int(prev_cursor_pos[1] * cursor_smoothing + cursor_y * (1 - cursor_smoothing))
                prev_cursor_pos = (smooth_x, smooth_y)
                pyautogui.moveTo(smooth_x, smooth_y, duration=0.05)
            
            # Get eyelid distances
            right_upper_lid = face_landmarks.landmark[159].y * img_h
            right_lower_lid = face_landmarks.landmark[145].y * img_h
            left_upper_lid = face_landmarks.landmark[386].y * img_h
            left_lower_lid = face_landmarks.landmark[374].y * img_h
            right_eye_distance = abs(right_upper_lid - right_lower_lid)
            left_eye_distance = abs(left_upper_lid - left_lower_lid)
            
            # Right-eye single blink for left click
            if right_eye_distance < blink_threshold and left_eye_distance >= blink_threshold:
                pyautogui.click(button='left')
                cv2.putText(frame, "LEFT CLICK!", (img_w//2, 50), 
                           cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 0, 255), 3)
            
            # Right-eye double blink for right click
            if right_eye_distance < blink_threshold:
                right_blink_times.append(time.time())
                right_blink_times = [t for t in right_blink_times if time.time() - t < blink_window]
                if len(right_blink_times) >= 2 and left_eye_distance >= blink_threshold:
                    pyautogui.click(button='right')
                    cv2.putText(frame, "RIGHT CLICK!", (img_w//2, 80), 
                               cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 255, 0), 3)
                    right_blink_times.clear()  # Reset after right click
            
            # Both eyes double blink for double click
            if right_eye_distance < blink_threshold and left_eye_distance < blink_threshold:
                both_blink_times.append(time.time())
                both_blink_times = [t for t in both_blink_times if time.time() - t < blink_window]
                if len(both_blink_times) >= 2:
                    pyautogui.doubleClick()
                    cv2.putText(frame, "DOUBLE CLICK!", (img_w//2, 110), 
                               cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255, 0, 0), 3)
                    both_blink_times.clear()  # Reset after double click
            
            # Display debug info with eyelid distances
            debug_text = (f"Eye Tracking Active | Smoothing: {cursor_smoothing*100:.0f}% | "
                         f"R Eye Dist: {right_eye_distance:.1f} | L Eye Dist: {left_eye_distance:.1f}")
            cv2.putText(frame, debug_text, (20, 30), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.7, (200, 255, 255), 1)

    # Add aesthetic frame decoration
    cv2.rectangle(frame, (0, 0), (img_w, img_h), (50, 50, 100), 2)
    cv2.putText(frame, "Gaze Control System", (img_w-250, 30), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (200, 200, 255), 1)
    
    cv2.imshow("Gaze Control", frame)
    if cv2.waitKey(1) == 27:  # ESC to quit
        break

cap.release()
cv2.destroyAllWindows()
```
