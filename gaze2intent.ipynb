```python
import cv2
import mediapipe as mp
import numpy as np
import pyautogui

# Initialize Mediapipe FaceMesh and Webcam
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(max_num_faces=1, refine_landmarks=True)

cap = cv2.VideoCapture(0)

# Constants
move_amount = 20  # pixels for mouse movement
blink_threshold = 5  # tune this as needed
screen_w, screen_h = pyautogui.size()

# Function to get iris center
def get_iris_center(landmarks, indices, w, h):
    x_list = [landmarks[idx].x * w for idx in indices]
    y_list = [landmarks[idx].y * h for idx in indices]
    center_x = int(sum(x_list) / len(x_list))
    center_y = int(sum(y_list) / len(y_list))
    return center_x, center_y

# Start Webcam Loop
while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    frame = cv2.flip(frame, 1)
    h, w, _ = frame.shape

    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = face_mesh.process(rgb_frame)

    if result.multi_face_landmarks:
        for face_landmarks in result.multi_face_landmarks:
            # Eye Corners (Green Dots)
            # left_eye_corners = [33, 133]
            # for idx in left_eye_corners:
            #     x = int(face_landmarks.landmark[idx].x * w)
            #     y = int(face_landmarks.landmark[idx].y * h)
            #     cv2.circle(frame, (x, y), 2, (0, 255, 0), -1)

            
            # Draw Green Dots on Both Eyes (Corners)
            eye_corners = [33, 133, 362, 263]  # Left + Right eye corners
            for idx in eye_corners:
                x = int(face_landmarks.landmark[idx].x * w)
                y = int(face_landmarks.landmark[idx].y * h)
                cv2.circle(frame, (x, y), 2, (0, 255, 0), -1)



            # Iris Centers (Red Dots)
            left_iris_indices = [474, 475, 476, 477]
            x_left, y_left = get_iris_center(face_landmarks.landmark, left_iris_indices, w, h)
            cv2.circle(frame, (x_left, y_left), 5, (0, 0, 255), -1)

            # Right Iris Centers (Red Dot)
            right_iris_indices = [469, 470, 471, 472]
            x_right, y_right = get_iris_center(face_landmarks.landmark, right_iris_indices, w, h)
            cv2.circle(frame, (x_right, y_right), 5, (0, 0, 255), -1)


            # PHASE 2: Gaze Detection3
            corner_left_x = int(face_landmarks.landmark[33].x * w)
            corner_right_x = int(face_landmarks.landmark[133].x * w)
            iris_ratio = (x_left - corner_left_x) / (corner_right_x - corner_left_x)

            if iris_ratio < 0.35:
                gaze_direction = "RIGHT"
                pyautogui.moveRel(move_amount, 0)
            elif iris_ratio > 0.65:
                gaze_direction = "LEFT"
                pyautogui.moveRel(-move_amount, 0)
            else:
                gaze_direction = "CENTER"

            cv2.putText(frame, f'Gaze: {gaze_direction}', (30, 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)

            # PHASE 3: Blink Detection + Click
            y_upper = int(face_landmarks.landmark[159].y * h)
            y_lower = int(face_landmarks.landmark[145].y * h)
            eye_distance = abs(y_upper - y_lower)

            cv2.putText(frame, f'Blink Dist: {eye_distance:.2f}', (30, 60),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)

            if eye_distance < blink_threshold:
                pyautogui.click()
                cv2.putText(frame, 'CLICKED!', (30, 90),
                            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Show Webcam Frame
    cv2.imshow('Gaze2Intent - Eye Gaze Mouse Control', frame)

    if cv2.waitKey(1) & 0xFF == 27:  # ESC to exit
        break

cap.release()
cv2.destroyAllWindows()
```
